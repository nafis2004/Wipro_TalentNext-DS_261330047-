1. Print 4th element from first and 4th element from last

```python
t = (10, 20, 30, 40, 50, 60, 70)

print("4th element from first:", t[3])
print("4th element from last:", t[-4])
```

2. Check whether an element exists in a tuple

```python
t = (10, 20, 30, 40, 50)

x = int(input("Enter element: "))

if x in t:
    print("Element exists")
else:
    print("Element does not exist")
```

3. Convert a list into a tuple

```python
lst = [10, 20, 30, 40, 50]

t = tuple(lst)

print("Tuple:", t)
```

4. Find the index of an item in a tuple

```python
t = (10, 20, 30, 40, 50)

x = int(input("Enter element: "))

if x in t:
    print("Index:", t.index(x))
else:
    print("Element not found")
```

5. Replace last value of tuples in a list to 100

```python
lst = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

result = []

for t in lst:
    new_tuple = t[:-1] + (100,)
    result.append(new_tuple)

print(result)
```
