5. Find longest word in file

```python
file = open("sample.txt", "r")
words = file.read().split()
file.close()

longest = max(words, key=len)

print("Longest word:", longest)
```

6. Count frequency of a user-entered word

```python
word = input("Enter word to search: ").lower()

file = open("sample.txt", "r")
data = file.read().lower()
file.close()

count = data.split().count(word)

print("Frequency of", word, "is", count)
```
