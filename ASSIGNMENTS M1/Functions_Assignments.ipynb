1. Sum of all numbers in a list

```python
def sum_list(lst):
    return sum(lst)

lst = [8, 2, 3, 0, 7]
print(sum_list(lst))
```

2. Reverse Of a String

```python
def reverse_string(s):
    return s[::-1]

s = "1234abcd"
print(reverse_string(s))
```

3. Factorial of a number

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))
```

4. Count Uppercase and Lowercase Letters

```python
def count_case(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

count_case("Hello World")
```

5. Print Even numbers from a list

```python
def even_numbers(lst):
    even = []

    for num in lst:
        if num % 2 == 0:
            even.append(num)

    return even

lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(even_numbers(lst))
```

6. Check whether a number is prime

```python
def is_prime(n):
    if n <= 1:
        return "Not Prime"

    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"

    return "Prime"

print(is_prime(7))
```
