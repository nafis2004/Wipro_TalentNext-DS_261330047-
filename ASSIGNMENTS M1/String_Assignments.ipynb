1. Count uppercase and lowercase letters in a string

```python
s = input("Enter a string: ")

upper = 0
lower = 0

for ch in s:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase letters:", upper)
print("Lowercase letters:", lower)
```

2. Check whether a string is palindrome or not

```python
s = input("Enter a string: ")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

3. Make a new string using first 2 characters repeated n times

```python
s = input("Enter a string: ")

first_two = s[:2]
result = first_two * len(s)

print(result)
```

4. Remove 'x' from first or last position

```python
s = input("Enter a string: ")

if s.startswith('x'):
    s = s[1:]

if s.endswith('x'):
    s = s[:-1]

print(s)
```

5. Repeat the last n characters n times

```python
s = input("Enter a string: ")
n = int(input("Enter n: "))

last_part = s[-n:]
result = last_part * n

print(result)
```
