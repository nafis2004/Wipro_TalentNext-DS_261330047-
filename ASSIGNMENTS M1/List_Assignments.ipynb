1. Create a List of 5 Integers & display the list items

```python
nums = [10, 20, 30, 40, 50]

print("Complete list:", nums)

for i in range(len(nums)):
    print("Index", i, "=", nums[i])
```

2. Append a new item to the end of the list

```python
nums = [10, 20, 30, 40]

nums.append(50)

print(nums)
```

3. Reverse the order of items in the list

```python
nums = [10, 20, 30, 40, 50]

nums.reverse()

print(nums)
```

4. Count occurences of a specified element

```python
nums = [10, 20, 30, 20, 40, 20]

x = int(input("Enter element: "))

count = nums.count(x)

print("Occurrences:", count)
```

5. Appends items of list1 to list2 in front

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = list1 + list2

print(result)
```

6. Insert a new element before second element

```python
nums = [10, 20, 30, 40]

nums.insert(1, 15)

print(nums)
```

7. Remove item from a specified index

```python
nums = [10, 20, 30, 40, 50]

index = int(input("Enter index to remove: "))

nums.pop(index)

print(nums)
```

8. Remove first occurence of a specified element

```python
nums = [10, 20, 30, 20, 40]

x = int(input("Enter element to remove: "))

nums.remove(x)

print(nums)
```
