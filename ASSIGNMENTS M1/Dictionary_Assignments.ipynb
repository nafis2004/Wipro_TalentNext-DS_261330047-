1. Add a key and value to a dictionary

```python
d = {0: 10, 1: 20}

d[2] = 30

print(d)
```

2. Concatenate multiple dictionaries

```python
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

result = {}
result.update(dic1)
result.update(dic2)
result.update(dic3)

print(result)
```

3. Check if a key exist

```python
d = {1: "Apple", 2: "Banana", 3: "Mango"}

key = int(input("Enter key: "))

if key in d:
    print("Key exists")
else:
    print("Key not found")
```

4. Iterate dictionary(Keys , Values, both)

```python
d = {1: "A", 2: "B", 3: "C"}

print("Keys:")
for key in d.keys():
    print(key)

print("Values:")
for value in d.values():
    print(value)

print("Key-Value pairs:")
for key, value in d.items():
    print(key, value)
```

5. Create dictionary from 1 to 15 with square values

```python
d = {}

for i in range(1, 16):
    d[i] = i * i

print(d)
```

6. Sum all values in dictionary

```python
d = {1: 10, 2: 20, 3: 30, 4: 40}

total = sum(d.values())

print("Sum =", total)
```
