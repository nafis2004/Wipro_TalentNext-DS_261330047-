1. Remove a given item from the set

```python
s = {10, 20, 30, 40, 50}

x = int(input("Enter element to remove: "))

if x in s:
    s.remove(x)
    print(s)
else:
    print("Element not found")
```

2. Create intersection of sets

```python
set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

result = set1.intersection(set2)

print("Intersection:", result)
```

3. Create union of sets

```python
set1 = {10, 20, 30}
set2 = {30, 40, 50}

result = set1.union(set2)

print("Union:", result)
```

4. Find maximum and minimum value in a set

```python
s = {15, 8, 25, 3, 19}

print("Maximum:", max(s))
print("Minimum:", min(s))
```
