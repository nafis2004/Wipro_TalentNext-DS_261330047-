Mini Project : **Purchase**

Creating a text file by the name(Purchase-1.txt) and writing items in the file

```python
with open("Purchase-1", "w") as file:
    file.write("""Chocolate 50
Biscuit 35
Icecream 50
Rice 100
Chicken 250

Perfume Free
Soup Free

Discount 80
""")
```

```python
try:
    filename = input("Enter the file name: ")

    file = open(filename, "r")
    lines = file.readlines()

    item_count = 0
    free_count = 0
    total = 0
    discount = 0

    for line in lines:
        line = line.strip()

        # Skip empty lines
        if line == "":
            continue

        parts = line.split()

        # Check if it is discount line
        if parts[0].lower() == "discount":
            discount = int(parts[1])

        else:
            item_name = parts[0]
            price = parts[1]

            # Check free items
            if price.lower() == "free":
                free_count += 1
            else:
                total += int(price)
                item_count += 1

    final_amount = total - discount

    print("No of items purchased:", item_count)
    print("No of free items:", free_count)
    print("Amount to pay:", total)
    print("Discount given:", discount)
    print("Final amount paid:", final_amount)

    file.close()

except FileNotFoundError:
    print("File not found")

except ValueError:
    print("Invalid data in file")

except Exception as e:
    print("Error:", e)
```
