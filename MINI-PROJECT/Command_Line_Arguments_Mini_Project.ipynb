Project : Happiness

```python
str1 = input("Enter first string: ").split("-")
str2 = input("Enter second string: ").split("-")
str3 = input("Enter third string: ").split("-")

happiness = 0

for num in str3:
    if num in str1:
        happiness += 1
    elif num in str2:
        happiness -= 1

print("Final happiness:", happiness)
```
