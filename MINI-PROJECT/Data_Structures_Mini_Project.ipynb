Project1. People and Facts (Dictionary)

```python
people = {
    "Jeff": "Is afraid of Dogs",
    "David": "Plays the piano",
    "Jason": "Can fly an airplane"
}
print("Original List:")
for name, fact in people.items():
    print(name + ":", fact)
people["Jeff"] = "Is afraid of heights"
people["Jill"] = "Can hula dance"
print("\nUpdated List:")
for name, fact in people.items():
    print(name + ":", fact)
```

Project2. Find the Runner Up Score(List)

```python
scores = [2, 3, 6, 6, 5]
unique_scores = list(set(scores))
unique_scores.sort()

print("Runner-up score:", unique_scores[-2])
```

Project3. Find the Percentage(Dictionary and List)

```python
students = {
    "Krishna": [67, 68, 69],
    "Arjun": [70, 98, 63],
    "Malika": [52, 56, 60]
}

name = input("Enter a name: ")

marks = students[name]

average = sum(marks) / len(marks)

print("Average percentage mark:", average)
```
